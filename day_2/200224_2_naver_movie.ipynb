```python
import datetime

days_range = []
```

```python
start = datetime.datetime.strptime("2019-10-1", "%Y-%m-%d")  # 수집 시작 날짜
print(start)
```

```python
end = datetime.datetime.strptime("2020-02-1", "%Y-%m-%d")  # 수집 종료 날짜 + 1
print(end)
```

```python
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end - start).days)]
print(date_generated)
```

```python
for date in date_generated:
    days_range.append(date.strftime("%Y-%m-%d"))
    print(date.strftime("%Y-%m-%d"))
```

```python
days_range
```

```python
import urllib
from tqdm import tqdm_notebook
import re

# movie_date = []
# movie_name = []
# movie_point = []

for today in tqdm_notebook(days_range):
    html = "http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=" + re.sub('-', '', str(today))
    response = urlopen(html)
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', class_='point'))
    
    movie_date.extend([today for n in range(end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie['date'] = movie['date'].astype('datetime64[ns]')
movie.head(20)
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
movie.info()
```

```python
import numpy as np

movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.mean)
# movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
# tmp = movie.query('name == ["82년생 김지영"]')
tmp = movie[movie['name'] == '포드 V 페라리']
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```
