```python
import time
from selenium import webdriver # import the webdriver of Selenium

# In order to start the Chrome browser, you need to specify the path of the folder containing the chrome driver
CHORME_PATH = '/Users/jehoonchae/Dropbox/Pycharm/chromedriver_mac64/chromedriver' # 자신의 크롬 드라이버 경로를 입력하세요
driver = webdriver.Chrome(executable_path = CHORME_PATH)
url = 'http://www.yes24.com/'
time.sleep(1)
driver.get(url) 
element = driver.find_element_by_xpath('//*[@id="yesFixCorner"]/dl/dd/ul[1]/li[1]/a') # xpath link of the elementt
element.click()
html = driver.page_source
time.sleep(3)
driver.close()
```
