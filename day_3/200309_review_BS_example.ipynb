### 3. NAVER 뉴스 스크랩 함수 작성 예제

```python
def get_article(url):
    
    # Obtain three types of information about a news article
#     r = requests.get(url)
    r = urlopen(url)
    soup = BeautifulSoup(r, 'lxml')
    news_title = soup.title.text
#     news_title = soup.find('title').text
    publisher = soup.find('meta', attrs={'name':'twitter:creator'}).get('content')
    news_content = soup.find('div', attrs = {'id':'articleBodyContents'}).text
    news_content = news_content.split('{}')[1].strip()
    
    return news_title, publisher, news_content
```

```python
url = 'https://news.naver.com/main/read.nhn?oid=421&sid1=101&aid=0003525799&mid=shm&mode=LSD&nh=20180808204753'

title, publisher, content = get_article(url)
```

```python
print(title)
print(publisher)
print(content)
```

```python
url = 'https://search.naver.com/search.naver?&where=news&query=%EC%BD%94%EB%A1%9C%EB%82%98&sm=tab_pge&sort=1&photo=0&field=0&reporter_article=&pd=0&ds=&de=&docid=&nso=so:dd,p:all,a:all&mynews=0&start=1&refresh_start=0'
```

```python
html = urlopen(url)
```

```python
html
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
soup
```

```python
article_list = soup.find('ul', {'class':'type01'}).find_all('li')
# class="type01"
```

```python
len(article_list)
```

```python
article_list[0].find('dd').find('a').get('href')
```

```python
url_list = []

for article in article_list:
    url = article.find('dd').find('a').get('href')
    if url == '#':
        continue
    url_list.append(url)
```
